## Python como Calculadora de Ahorros üí∞
¬°El nuevo trabajo de ciencia de datos le ha dejado mucho dinero para invertir!

Usemos python para ver lo que puedes ganar invirtiendo en un banco.


```python
ingresos = []
egresos = []

while True:
    tipo = input("Ingresa 'I' para a√±adir un ingreso, 'E' para a√±adir un egreso o 'S' para salir: ").upper()

    if tipo == 'I':
        monto = float(input("Ingresa el monto del ingreso: "))
        ingresos.append(monto)
        print(f"Ingreso de ${monto} a√±adido.")
    elif tipo == 'E':
        monto = float(input("Ingresa el monto del egreso: "))
        egresos.append(monto)
        print(f"Egreso de ${monto} a√±adido.")
    elif tipo == 'S':
        break
    else:
        print("Opci√≥n no v√°lida. Intenta nuevamente.")

total_ingresos = sum(ingresos)
total_egresos = sum(egresos)
balance = total_ingresos - total_egresos

print("\n--- Resumen de Transacciones ---")
print(f"Ingresos totales: ${total_ingresos:.2f}")
print(f"Egresos totales: ${total_egresos:.2f}")
print(f"Balance final: ${balance:.2f}")

if balance > 0:
    print("Tienes un balance positivo. ¬°Bien hecho!")
elif balance < 0:
    print("Tienes un balance negativo. Revisa tus egresos.")
else:
    print("Tu balance es neutro, sin ganancias ni p√©rdidas.")

print("\n--- Detalle de Transacciones ---")
print("Ingresos:")
for ingreso in ingresos:
    print(f" + ${ingreso:.2f}")
print("Egresos:")
for egreso in egresos:
    print(f" - ${egreso:.2f}")
```

```python
# Su banco est√° ofreciendo 7.01%
# (Inter√©s Simple, por a√±o)
# Asignar los valores num√©ricos solamente (sin el %)
# p.ej. si la tasa de inter√©s es 7.01%, asigne el valor como 7.01/100

interest_rate = float(input("Ingresa la tasa de inter√©s (en porcentaje): ")) / 100
```

```python
# Veamos cu√°nto ganar√≠a despu√©s de 2 a√±os
num_years = int(input("Ingresa el n√∫mero de a√±os de la inversi√≥n: "))
# Ahorros totales = Monto de capital + (Monto de capital x Tasa de inter√©s x N√∫mero de a√±os)
# Sugerencia: consulte la descripci√≥n del ejercicio para ver la f√≥rmula
# Usando los nombres de las variables asignadas, imprima la nueva cantidad
#Encuentra el inter√©s ganado despu√©s de un n√∫mero dado de a√±os
nuevo_monto = ahorros + (ahorros * interest_rate * num_years)
if nuevo_monto > 1200:
    print("¬°Tu inversi√≥n est√° rindiendo bien!")
else:
    print("Considera invertir en opciones con mayor inter√©s.")
print(f'El inter√©s total ganado por una inversi√≥n de {num_years} a√±os es ${nuevo_monto:.2f}')
```
